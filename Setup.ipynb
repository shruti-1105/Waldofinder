```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
import pandas as pd
import numpy as np
from PIL import Image
import io
import os
```

We first read in the images as numpy arrays.

```python
w = np.array([np.array(imread("../Hey-Waldo/images/64/waldo/"+fname)) for fname in os.listdir("../Hey-Waldo/images/64/waldo")])
nw = np.array([np.array(imread("../Hey-Waldo/images/64/notwaldo/"+fname)) for fname in os.listdir("../Hey-Waldo/images/64/notwaldo")])
```

Let's take a look at the data.

```python
plt.imshow(w[0])
```

To clean our data up, we flatten the arrays and save to a csv

```python
data = []
for im in w:
    data.append(im.flatten('F'))
```

```python
df = pd.DataFrame(data)
df['waldo'] = 1
```

```python
# df.to_csv('waldo128.csv',index=False)
```

```python
data = []
for im in nw:
    data.append(im.flatten('F'))
```

```python
df2 = pd.DataFrame(data)
df2['waldo'] = 0
```

```python
# df2.to_csv('notwaldo128.csv',index=False)
```

```python
frames = [df, df2]

result = pd.concat(frames)
```

```python
result.to_csv('all_waldo64.csv',index=False)
```

We can easily see that our transformation preserves the original data.
